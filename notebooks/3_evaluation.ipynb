```python
from PIL import Image
import numpy as np
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
# Path to store predictions
predictions_path = "/Users/mariarisques/TAED2_YOLOs/data/predictions/"
folder_path = os.path.join('/Users/mariarisques/TAED2_YOLOs/data/interim/transformed/images/val/')
file_names = os.listdir(folder_path)
file_names = [folder_path + file for file in file_names if os.path.isfile(str(os.path.join(folder_path, file)))]
```

```python
mIoU = 0

for idx, file in enumerate(file_names):
  try:
    file_name = file.split('/')[-1]
    true_mask = np.asarray(Image.open(file.replace('images', 'masks').replace('jpg', 'png')))
    pred_mask = np.asarray(Image.open(predictions_path + file_name))

    gtb = (true_mask > 0)
    predb = (pred_mask > 0)

    overlap = gtb * predb
    union = gtb + predb
    IoU = overlap.sum() / union.sum()

    print(f"{idx}. IoU for file {file_name} is: {IoU}")

    mIoU += IoU
  except Exception as e:
    print(f'Could not compute mIoU for file {file_name}')
    print(f'Error: {e}')

mIoU /= len(file_names)
print(f"Global mIoU = {mIoU:0.4f}")
```
